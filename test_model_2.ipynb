```python
import pandas as pd
data = pd.read_csv('/content/housing_price_dataset.csv')
```

```python
data.head()
```

```python
data['SquareFeet'] = (data['SquareFeet'] - data['SquareFeet'].mean())/(data['SquareFeet'].max() - data['SquareFeet'].min())
```

```python
data.head()
```

```python
data['Bedrooms'] = (data['Bedrooms'] - data['Bedrooms'].mean())/(data['Bedrooms'].max() - data['Bedrooms'].min())
```

```python
data['Bathrooms'] = (data['Bathrooms'] - data['Bathrooms'].mean())/(data['Bathrooms'].max() - data['Bathrooms'].min())
```

```python
df = pd.DataFrame(data)
```

```python
df = pd.get_dummies(df, columns=['Neighborhood'], dtype=int)
```

```python
df
```

```python
import torch

y = torch.tensor(df['Price'].values)
```

```python
y
```

```python
y = (y - y.mean())/(y.max() - y.min())
```

```python
df = df.drop(columns = ['Price'])
```

```python
df
```

```python
data['YearBuilt'] = (data['YearBuilt'] - data['YearBuilt'].mean())/(data['YearBuilt'].max() - data['YearBuilt'].min())
```

```python
df['YearBuilt'] = (df['YearBuilt'] - df['YearBuilt'].mean())/(df['YearBuilt'].max() - df['YearBuilt'].min())
```

```python
df
```

```python
X = torch.tensor(df.values)
```

```python
X
```

```python
X.shape
```

```python
train_ind = int(0.8 * len(X))
```

```python
train_ind
```

```python
X_train, y_train, X_test, y_test = X[:train_ind], y[:train_ind], X[train_ind:], y[train_ind:]
```

```python
len(X_train)
```

```python
from torch import nn
```

```python
class LinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer = nn.Linear(in_features = 7,
                           out_features = 1)

  def forward(self, x: torch.Tensor) -> torch.Tensor:
    return self.layer(x)
```

```python
torch.manual_seed(42)
model = LinearRegressionModel()
```

```python
model.state_dict()
```

```python
from torch import optim
loss = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.001)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1000, gamma=0.05)
```

```python
model.state_dict()

```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)

X_test_batch = X_test[:1000].to(device).to(torch.float32)
y_test_batch = y_test[:1000].to(device).to(torch.float32)
X_train_batch = X_train[:10000].to(device).to(torch.float32)
y_train_batch = y_train[:10000].to(device).to(torch.float32)
```

```python
device
```

```python
X_train.device
```

```python
import torch

torch.cuda.empty_cache()
torch.cuda.reset_peak_memory_stats()
```

```python
epochs = 10000

loss_hist = []
ep_num = []
te = []

for epoch in range(epochs + 1):
  model.train()
  pred = model(X_train_batch)
  loss_value = loss(pred, y_train_batch)
  optimizer.zero_grad()
  loss_value.backward()
  optimizer.step()
  scheduler.step()
  model.eval()

  with torch.no_grad():
    test_pred = model(X_test_batch)
    loss_test = loss(test_pred, y_test_batch)
  if epoch%100 == 0:
    ep_num.append(epoch)
    loss_hist.append(loss)
    te.append(loss_test)
    print(f"Epoch : {epoch} | Loss : {loss_value} | Test Loss : {loss_test}")
```

```python
model.state_dict()
```

```python
import matplotlib.pyplot as plt

# Get model predictions
y_pred = model(X_test.to(torch.float)).cpu().detach().numpy()  # Convert tensor to NumPy
y_actual = y_test.cpu().numpy()  # Convert actual values

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(y_actual, y_pred, alpha=0.6)
plt.plot([min(y_actual), max(y_actual)], [min(y_actual), max(y_actual)], 'r', linestyle="--")  # 45-degree line
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted Prices")
plt.show()
```
